```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
```

```python
fig1 = plt.figure(figsize=(4.5,4))
dropout_p = [0.6, 0.7, 0.85, 0.97]
train_error = np.array([0.549, 0.447, 0.329, 0.244])
valid_error = np.array([0.593, 0.508, 0.434, 0.457])
valid_acc = [0.807, 0.828, 0.851, 0.854]

ax = fig1.add_subplot(111)
ax.plot(dropout_p, valid_acc, 'r', label = 'Val. Acc.')
ax.set_xlabel('Dropout value')
ax.set_ylabel('Accuracy')
ax.set_ybound(0, 1)


ax2 = ax.twinx()
ax2.plot(dropout_p, valid_error - train_error, 'b', label = 'Gap')
print(valid_error-train_error)
ax2.set_ylabel('Generalization gap')
ax2.set_ybound(0, 1)
fig1.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)
plt.tight_layout()
fig1.savefig('dropout.pdf')
```

```python
fig2 = plt.figure(figsize=(4.5,4))
weight = [float(5e-4), float(1e-3), float(5e-3), float(5e-2)]
train_error_L1 = np.array([0.642, 0.841, 3.850, 3.850])
valid_error_L1= np.array([0.658, 0.849, 3.850, 3.850])
valid_acc_L1 = [0.795, 0.751, 0.0241, 0.0220]
train_error_L2 = np.array([0.306, 0.361, 0.586, 2.258])
valid_error_L2= np.array([0.460, 0.456, 0.607, 2.256])
valid_acc_L2 = [0.851, 0.850, 0.813, 0.392]

ax = fig2.add_subplot(111)
ax.plot(weight, valid_acc_L1, 'r', label = 'L1 Val. Acc.')
ax.plot(weight, valid_acc_L2, 'b', label = 'L2 Val. Acc.')
ax.set_xlabel('Weight decay value')
ax.set_ylabel('Accuracy')
ax.set_ybound(0, 1)

ax2 = ax.twinx()
ax2.plot(weight, valid_error_L1 - train_error_L1, 'r--', label = 'L1 Gap')
ax2.plot(weight, valid_error_L2 - train_error_L2, 'b--', label = 'L2 Gap')
print("L1 gap: ", valid_error_L1 - train_error_L1)
print("L2 gap: ", valid_error_L2 - train_error_L2)
ax2.set_ylabel('Generalization gap')
ax2.set_ybound(-0.01, 1)
fig2.legend(loc="upper right", bbox_to_anchor=(1,1), bbox_transform=ax.transAxes)
ax2.set_xscale('log')
plt.tight_layout()
fig2.savefig('weight_penalty.pdf')
```
